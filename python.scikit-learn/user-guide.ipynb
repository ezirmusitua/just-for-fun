##### Classification & Clustering  

* make_blobs  
    multiclass datasets, provides greater control regarding the centers and standard deviations of each cluster, and is used to demonstrate clustering

```python
datasets.make_blobs
```

* make_classification  
    specialises in introducing noise by way of: correlated, redundant and uninformative features; multiple Gaussian clusters per class; and linear transformations of the feature space  

```python
datasets.make_classification
```

* make_gaussian_quantiles  
     divides a single Gaussian cluster into near-equal-size classes separated by concentric hyperspheres  
     

```python
datasets.make_gaussian_quantiles
```

* make_hastie_10_2  
    generates a similar binary, 10-dimensional problem  
     

```python
datasets.make_hastie_10_2
```

* make_circles  
    2d binary classification datasets that are challenging to certain algorithms, useful for visualisation, produces Gaussian data with a spherical decision boundary for binary classification  

```python
datasets.make_circles
```

* make_moons  
    2d binary classification datasets that are challenging to certain algorithms, useful for visualisation, produces Gaussian data with a spherical decision boundary for binary classification  

```python
datasets.make_moons
```
